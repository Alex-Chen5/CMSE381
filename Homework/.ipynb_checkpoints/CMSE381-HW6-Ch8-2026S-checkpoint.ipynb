```python
url = "https://msu-cmse-courses.github.io/CMSE381-S26/_downloads/73b56f6984db80ab92105d5e86279f38/OJ.csv"
OJ_df = pd.read_csv(url, index_col=0)

# Convert the Store7 column from Yes/No to 1/0 so scikit-learn can use it.
if 'Store7' in OJ_df.columns:
    OJ_df['Store7'] = OJ_df['Store7'].replace({'Yes': 1, 'No': 0})

# Preview the data
print(OJ_df.info())

OJ_df.head()
```

## Question 1. Classification Tree on the OJ Data Set

### (a) (6 points)
Create a training set containing a random sample of **800 observations** and a test set containing the remaining observations.

Report:
- the percentage of observations in the training set



```python
####YOUR CODE HERE###

np.random.seed(42)
train_indices = np.random.choice(OJ_df.index, size=800, replace=False)
train_df = OJ_df.loc[train_indices]
test_df = OJ_df.drop(train_indices)
train_percent = len(train_df) / len(OJ_df) * 100

print(f"Training set size: {len(train_df)}")
print(f"Test set size: {len(test_df)}")
print(f"Percentage in training set: {train_percent:.2f}%")
```

### (b) (6 points)
Fit a **classification tree** to the training data, with **Purchase** as the response and all remaining variables as predictors. To control model complexity, we restrict the tree to have a maximum depth of 3.

Report:
- the fitted model
- the training error rate


```python
###YOUR CODE HERE###

X_train = train_df.drop('Purchase', axis=1)
y_train = train_df['Purchase']
clf = DecisionTreeClassifier(max_depth=3, random_state=42)
clf.fit(X_train, y_train)
y_train_pred = clf.predict(X_train)

train_error = 1 - accuracy_score(y_train, y_train_pred)
print(f"Training Error Rate: {train_error:.4f}")

plt.figure(figsize=(15,8))
plot_tree(clf, feature_names=X_train.columns, class_names=clf.classes_, filled=True)
plt.show()

# The fitted classification tree, with max depth = 3, is shown above, along with the training error rate of 0.1913
```

### (c) (3 + 3 points): Create a plot of the fitted tree and interpret the results.
1. Create a plot of the fitted tree.



```python
###YOUR CODE HERE###

plt.figure(figsize=(15,8))
plot_tree(clf, feature_names=X_train.columns, class_names=clf.classes_, filled=True)
plt.show()
```

### (c)
2. Interpret the results based on the above plot.

Answer the following:
- How many terminal nodes does the tree have?
- Which variables appear near the top of the tree?
- Briefly interpret two of the main splits.

###YOUR ANSWER HERE###

The tree has 8 terminal nodes.

The variable LoyalCH shows up at the very top of the tree and is used in the first few splits, which means it is the most important factor in predicting purchases. Other variables like PriceDiff and SalePriceMM show up further down, so they are a smaller role.

The first split is based on whether LoyalCH $ \leq 0.483$. This means that customers with lower loyalty to CH are more likely to buy MM, while those with higher loyalty usually stick with CH.

Another main split is PriceDiff $ \leq -0.39$. This suggests that even when MM is cheaper than CH, customers who are strongly loyal to CH still tend to choose CH.

```python
###YOUR CODE HERE###

X_test = test_df.drop('Purchase', axis=1)
y_test = test_df['Purchase']
y_test_pred = clf.predict(X_test)
cm = confusion_matrix(y_test, y_test_pred)
print("Confusion Matrix:")
print(cm)

test_error = 1 - accuracy_score(y_test, y_test_pred)
print(f"Test Error Rate: {test_error:.4f}")
```

## Question 2. Cross-Validation and Tree Size

### (a) (12 points)

Use cross-validation on the training set to determine an appropriate tree size. Vary the maximum depth of the tree over a range of 1 to 20 and use 5-fold cross-validation to evaluate model performance. Plot **tree size** versus **cross-validated classification error rate**.


```python
###YOUR CODE HERE##

depths = range(1, 21)
cv_errors = []
for d in depths:
    clf = DecisionTreeClassifier(max_depth=d, random_state=42)
    scores = cross_val_score(clf, X_train, y_train, cv=5)
    error = 1 - scores.mean()
    cv_errors.append(error)
plt.figure(figsize=(8,5))
plt.plot(depths, cv_errors, marker='o')
plt.xlabel("Tree Depth")
plt.ylabel("Cross-Validated Error Rate")
plt.title("Tree Depth vs Cross-Validated Error")
plt.show()
```

## Question 3. Pruned vs. Unpruned Tree


### (a) (10 points)

Fit both an **unpruned tree** and a **pruned (or restricted) tree** using the optimal tree size identified in Question 2. If the cross-validation results do not clearly support pruning, fit a smaller reasonable tree instead and briefly justify your choice.

For each model, report both the **training error rate** and the **testing error rate**, as these will be used for comparison in part (b) and (c).

Be sure to clearly state the chosen tree size (e.g., depth or number of terminal nodes) for the pruned/restricted model.

```python
###YOUR CODE HERE###

unpruned_clf = DecisionTreeClassifier(random_state=42)
unpruned_clf.fit(X_train, y_train)

y_train_pred_unpruned = unpruned_clf.predict(X_train)
train_error_unpruned = 1 - accuracy_score(y_train, y_train_pred_unpruned)

y_test_pred_unpruned = unpruned_clf.predict(X_test)
test_error_unpruned = 1 - accuracy_score(y_test, y_test_pred_unpruned)

print("Unpruned Tree:")
print(f"Training Error: {train_error_unpruned:.4f}")
print(f"Test Error: {test_error_unpruned:.4f}")


pruned_clf = DecisionTreeClassifier(max_depth=4, random_state=42)
pruned_clf.fit(X_train, y_train)

y_train_pred_pruned = pruned_clf.predict(X_train)
train_error_pruned = 1 - accuracy_score(y_train, y_train_pred_pruned)

y_test_pred_pruned = pruned_clf.predict(X_test)
test_error_pruned = 1 - accuracy_score(y_test, y_test_pred_pruned)

print("\nPruned Tree (max_depth = 4):")
print(f"Training Error: {train_error_pruned:.4f}")
print(f"Test Error: {test_error_pruned:.4f}")

# The pruned tree uses max depth = 4, based on cross-validation results.
```

### (b) (3 points)
Compare the **training error rates** of the pruned and unpruned trees. Which is higher?

###YOUR ANSWER HERE##

The pruned tree has the higher training error rate. The unpruned tree has a training error of 0.0112, while the pruned tree (max depth = 4) has a higher training error of 0.1587.

### (c) (3 points)
Compare the **test error rates** of the pruned and unpruned trees. Which model performs better?

###YOUR ANSWER HERE##

The pruned tree performs better on the test set. The pruned tree has a lower test error rate of 0.1667, compared to 0.2037 for the unpruned tree. This means that the pruned tree generalizes better to new data, and the unpruned tree probably overfits the training data.

### (d) (4 points)
Briefly explain why the pruned and unpruned trees may perform differently on training and test data.


###YOUR ANSWER HERE###

The pruned and unpruned trees perform differently because of overfitting. The unpruned tree is more complex and fits the training data very closely, so it has a very low training error. But it also picks up random noise and details that don’t apply to new data, which leads to a higher test error.

The pruned tree is simpler, so it does not fit the training data as perfectly and has a higher training error. But, because it focuses on the main patterns instead of noise, it does a better job on the test data.

## Question 5. Reflection (10 points)

Write a short paragraph addressing the following:

1. What is one advantage of a single decision tree?
2. What is one disadvantage of a single decision tree?
3. Why can ensemble methods such as random forests often improve prediction performance?


###YOUR ANSWER HERE###

A single decision tree has the advantage of being easy to understand and interpret, since you can clearly see how decisions are made based on the splits. But, a disadvantage is that decision trees can easily overfit the training data, especially if they are very deep, which can hurt performance on new data. Ensemble methods like random forests improve prediction performance by combining many trees together, which helps reduce overfitting and makes the model more stable and accurate overall.
